# A good developer must be a good detective

There is a bug in this program, how can we find it?

```python
def maximum(numbers):
    #pass # pass means "do nothing", add your code here
    max_value = 0
    for num in numbers:
        if num > max_value:
            max_value = num
    return max_value

def main():
    print("Let us find the maximum value from the following list:")
    list_of_nums = [1,2,3]
    max_value = maximum(list_of_nums)
    print(max_value)

    print("Let us find the maximum value from another list:")
    list_of_nums = [-1, -2, -3]
    max_value = maximum(list_of_nums)
    print(max_value)
    
main()
```

### Print internal values, as you execute the program

```python
def maximum(numbers):
    #pass # pass means "do nothing", add your code here
    max_value = 0
    for num in numbers:
        # print(num) # <= Don't just print the number, add some context around it
        print(f"Current max_value {max_value}")
        print(f"Current number {num}")
        if num > max_value:
            max_value = num
    return max_value

def main():
    print("Let us find the maximum value from the following list:")
    list_of_nums = [1,2,3]
    max_value = maximum(list_of_nums)
    print(max_value)

    print("Let us find the maximum value from another list:")
    list_of_nums = [-1, -2, -3]
    max_value = maximum(list_of_nums)
    print(max_value)
    
main()
```
